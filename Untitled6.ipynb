```python
from scipy.stats import shapiro

data=[1.2,2.5,3.1,4.2,5.8,6.9,7.5,8.6,9.1]

stat, p=shapiro(data)

print('Shapiro-wilk test statistic:', stat)
print('p-value', p)

alpha=0.05
if p> alpha:
    print('The data is normally distributed (fail to reject H0)')
else:
    print('The data is normally distributed (rejected H0)')
```
