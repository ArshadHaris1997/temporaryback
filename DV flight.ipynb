```python
import pandas as pd
```

```python
df_cleaned=pd.read_csv('synthetic_flight_passenger_data.csv')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set style
sns.set(style="whitegrid")

# Create the plot
plt.figure(figsize=(10, 6))
status_plot = sns.countplot(
    data=df_cleaned,
    x='Airline',
    hue='Flight_Status',
    palette='Set2'
)
plt.title('Flight Status Distribution Across Airlines')
plt.xlabel('Airline')
plt.ylabel('Number of Flights')
plt.legend(title='Flight Status')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
sns.barplot(
    data=df_cleaned,
    x='Seat_Class',
    y='Flight_Satisfaction_Score',
    estimator='mean',
    palette='coolwarm'
)

plt.title('Average Flight Satisfaction Score by Seat Class')
plt.xlabel('Seat Class')
plt.ylabel('Average Satisfaction Score')
plt.ylim(0, 10)  # scores range from 0 to 10
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.histplot(
    data=df_cleaned,
    x='Age',
    hue='Travel_Purpose',
    multiple='stack',
    bins=30,
    palette='Set3'
)

plt.title('Passenger Age Distribution by Travel Purpose')
plt.xlabel('Age')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df_cleaned,
    x='Distance_Miles',
    y='Price_USD',
    hue='Seat_Class',
    palette='viridis',
    alpha=0.6
)

plt.title('Ticket Price vs Flight Distance by Seat Class')
plt.xlabel('Distance (Miles)')
plt.ylabel('Price (USD)')
plt.legend(title='Seat Class')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df_cleaned,
    x='Departure_Delay_Minutes',
    y='Flight_Satisfaction_Score',
    hue='Flight_Status',
    palette='coolwarm',
    alpha=0.6
)

plt.title('Impact of Departure Delays on Passenger Satisfaction')
plt.xlabel('Departure Delay (Minutes)')
plt.ylabel('Satisfaction Score')
plt.legend(title='Flight Status')
plt.axvline(x=0, color='grey', linestyle='--')  # On-time cutoff
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.boxplot(
    data=df_cleaned,
    x='Travel_Purpose',
    y='Booking_Lead_Time_Days',
    hue='Flight_Status',
    palette='Set2'
)

plt.title('Booking Lead Time by Travel Purpose and Flight Status')
plt.xlabel('Travel Purpose')
plt.ylabel('Booking Lead Time (Days)')
plt.legend(title='Flight Status')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
sns.violinplot(
    data=df_cleaned,
    x='Airline',
    y='Distance_Miles',
    palette='pastel',
    inner='box'
)

plt.title('Distribution of Flight Distances by Airline')
plt.xlabel('Airline')
plt.ylabel('Distance (Miles)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create age bins
age_bins = [0, 18, 30, 45, 60, 75, 100]
age_labels = ['<18', '18-30', '31-45', '46-60', '61-75', '75+']
df_cleaned['Age_Group'] = pd.cut(df_cleaned['Age'], bins=age_bins, labels=age_labels)

# Plot
plt.figure(figsize=(10, 6))
sns.countplot(
    data=df_cleaned,
    x='Age_Group',
    hue='Checkin_Method',
    palette='Set1'
)

plt.title('Check-in Method Usage by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Passenger Count')
plt.legend(title='Check-in Method')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate no-show rates
no_show_rates = df_cleaned.groupby(['Travel_Purpose', 'Seat_Class'])['No_Show'].mean().reset_index()
no_show_rates['No_Show_Percentage'] = no_show_rates['No_Show'] * 100

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(
    data=no_show_rates,
    x='Seat_Class',
    y='No_Show_Percentage',
    hue='Travel_Purpose',
    palette='Set2'
)

plt.title('No-Show Rate by Travel Purpose and Seat Class')
plt.ylabel('No-Show Rate (%)')
plt.xlabel('Seat Class')
plt.legend(title='Travel Purpose')
plt.tight_layout()
plt.show()
```
